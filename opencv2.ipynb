```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread("/content/eyes.jpg")
cv2_imshow(img)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
re=cv2.resize(img,(500,200))
cv2_imshow(re)
```

```python
cir=cv2.circle(img,(250,80),60,(250,0,0),4)
cv2_imshow(cir)
```

```python
import numpy as np
import matplotlib.pyplot as plt
(h,w)=img.shape[:2]
center=(w//2,h//2)
angle=45
scale=1.0
rotation_matrix=cv2.getRotationMatrix2D(center,angle,scale)
rotated=cv2.warpAffine(img,rotation_matrix,(w,h))
plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.title("rotated Image(45)")
```

```python
flip=cv2.flip(img,1)
plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(flip,cv2.COLOR_BGR2RGB))
plt.title("Flipped img")
```

```python
tx,ty=20,30
translation_matrix=np.float32(([1,0,tx],[0,1,ty]))
translated=cv2.warpAffine(img,translation_matrix,(w,h))
plt.subplot(2,3,2)
plt.imshow(cv2.cvtColor(translated,cv2.COLOR_BGR2RGB))
plt.title("Translated img")
```
